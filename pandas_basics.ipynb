**Handling missing data**

```python
df2 = pd.DataFrame({
    'Name':['Alice','Bob','Charlie',None],
    'Age':[24,None,22,32],
    'Score':[85,95,None,88]
})
```

```python
df2.head()
```

```python
df2.dropna(subset=["Name"],inplace=True)
```

```python
df2.head()
```

```python
df2["Age"] = df2["Age"].fillna(df2["Age"].mean())
```

```python
df2.head()
```

```python
df2["Score"]  = df2["Score"].fillna(df2["Score"].mean())
```

```python
df2.head()
```
