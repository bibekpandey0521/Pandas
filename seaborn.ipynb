```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.__version__)
```

**Load in-built seaborn dataset (tips)**

```python
tips = sns.load_dataset("tips")
```

```python
tips.shape
```

```python
tips.head()
```

```python
tips.tail()
```

```python
tips["day"].unique()
```

```python
tips[tips["day"] == "Sun"]
sun = tips[tips["day"] == "Sun"]

mean_bill = sun["total_bill"].mean()
median_bill = sun["total_bill"].median()

print("Mean:", mean_bill)
print("Median:", median_bill)
```

```python
# set global styles and color palette
sns.set_style("darkgrid")
sns.set_palette("viridis")
```

```python
plt.figure(figsize=(6,4))
sns.histplot(tips["total_bill"],kde=True,color="teal")
plt.title("Distribution of Total bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

**Box Plot - Detect Outliers**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="day",y="total_bill",data=tips,palette="pastel",legend=False)
plt.title("Total Bill by day")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="sex",y="total_bill",data=tips,palette="pastel",legend=False)
plt.title("Total Bill by day")
plt.show()
```

**Count Plot - number of records per day**

```python
plt.figure(figsize=(6,4))
sns.countplot(x="day",data=tips,hue="sex")
plt.title("Total Bill by day")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex",size="size",palette="cool")
plt.title("Number of Record per day")
plt.show()
```

**Pair plot - Pair wise Feature realtionship**

```python
sns.pairplot(tips,hue="sex",diag_kind="kde")
plt.suptitle("Pair-wise relationship")
plt.show()
```

**HeatMap-Feature Corelation**

```python
correlation_matrix = tips.corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",linewidths=0.5)
plt.title("Feature of Corealation Heatmap")
plt.show()
```
