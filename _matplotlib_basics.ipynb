## Finding the spacing

`np.linspace(0, 10, 50)` creates 50 evenly spaced points between 0 and 10 (inclusive).

Since there are 50 points, there are 49 intervals between them. The spacing between consecutive points is

$\frac{10 - 0}{50 - 1} = \frac{10}{49} \approx 0.20408163$

Therefore, each value in the array is approximately 0.20408163 greater than the previous value.

## Example

```python
import numpy as np

x = np.linspace(0, 10, 50)
print(x)
```

# Understanding the 3rd Value in `np.linspace(0, 10, 50)`

```python
import numpy as np

x = np.linspace(0, 10, 50)
```

## What Does `linspace()` Do?

`np.linspace(start, stop, num)` generates **`num` evenly spaced values** between `start` and `stop`, including both endpoints.

For this example:

- `start = 0`
- `stop = 10`
- `num = 50`

---

## Step 1: Compute the Step Size

The spacing between consecutive values is:

$$
\text{step} = \frac{\text{stop} - \text{start}}{\text{num} - 1}
$$

Substituting the values:

$$
\text{step}
= \frac{10 - 0}{50 - 1}
= \frac{10}{49}
\approx 0.20408163
$$

---

## Step 2: Compute Each Value

Each value is given by:

$$
x[i] = \text{start} + i \cdot \text{step}
$$

where `i` is the zero-based index.

---

## First Three Values

| Position | Index | Value |
|----------|--------|--------|
| 1st | `0` | 0.00000000 |
| 2nd | `1` | 0.20408163 |
| 3rd | `2` | 0.40816327 |

---

## Calculating the 3rd Value

The 3rd value corresponds to index `2`.

$$
x[2]
= 0 + 2 \times \frac{10}{49}
$$

$$
= \frac{20}{49}
$$

$$
\approx 0.40816327
$$

Therefore:

```python
x[2]
```

returns:

```text
0.40816326530612246
```

---

## Verification

```python
import numpy as np

x = np.linspace(0, 10, 50)

print(x[2])
```

### Output

```text
0.40816326530612246
```

---

## Why `x[2]` and Not `x[3]`?

Python uses **zero-based indexing**:

| Value Position | Python Index |
|---------------|--------------|
| 1st value | `x[0]` |
| 2nd value | `x[1]` |
| 3rd value | `x[2]` |
| 4th value | `x[3]` |

So the 3rd value is accessed using:

```python
x[2]
```

---

## Visual Check

```python
print(x[:5])
```

Output:

```text
[0.         0.20408163 0.40816327 0.6122449  0.81632653]
```

Notice that the third element in the array is:

```text
0.40816327
```

---

## General Formula

For:

```python
x = np.linspace(start, stop, num)
```

the step size is:

$$
\text{step} = \frac{\text{stop} - \text{start}}{\text{num} - 1}
$$

and the value at index `i` is:

$$
x[i] = \text{start} + i \cdot \text{step}
$$

For this example:

$$
x[2]
= 0 + 2 \times \frac{10}{49}
\approx 0.40816327
$$

✅ **The 3rd value in `np.linspace(0, 10, 50)` is approximately `0.40816327`.**

**Simple Line Plot**

````python
x = np.linspace(0,10,50)
y = np.sin(x)
plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y = sin(X)")
plt.show()
````

**Customized Plot**

````python
plt.figure(figsize=(7,4))
plt.plot(x,y,color="red",linestyle="--", marker='o',label="sin(x)")
plt.title("Customized Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
````

**Bar Chart**

**Multiple Plots(sub-plots)**

````python
x = np.linspace(0,10,100)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(x,np.sin(x),"r--")
plt.title("Sine")

plt.subplot(1,2,2)
plt.plot(x,np.cos(x),"b--")
plt.title("Cosine")

plt.suptitle("Sine and cosine functions")
plt.show()
````
