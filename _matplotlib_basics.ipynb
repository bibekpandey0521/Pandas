## Finding the spacing

`np.linspace(0, 10, 50)` creates 50 evenly spaced points between 0 and 10 (inclusive).

Since there are 50 points, there are 49 intervals between them. The spacing between consecutive points is

$\frac{10 - 0}{50 - 1} = \frac{10}{49} \approx 0.20408163$

Therefore, each value in the array is approximately 0.20408163 greater than the previous value.

`np.linspace(0, 10, 50)` generates 50 evenly spaced values between 0 and 10.

$\frac{10 - 0}{50 - 1} = \frac{10}{49} \approx 0.20408163$

The first value is 0, the last value is 10, and every adjacent pair differs by approximately 0.20408163.

**Simple Line Plot**

````python
x = np.linspace(0,10,50)
y = np.sin(x)
plt.plot(x,y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y = sin(X)")
plt.show()
````

**Customized Plot**

````python
plt.figure(figsize=(7,4))
plt.plot(x,y,color="red",linestyle="--", marker='o',label="sin(x)")
plt.title("Customized Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.grid(True)
plt.show()
````
